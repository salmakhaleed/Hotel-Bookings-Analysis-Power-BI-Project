```python
import pandas as pd
pd.__version__
```

```python
df = pd.read_csv("hotels.csv")
df.head()
```

```python
df.columns
```

```python
df['booking_id'] = range(1, len(df) + 1)
df.head()
```

```python
dim_hotel = df[['hotel']].drop_duplicates().reset_index(drop=True)
dim_hotel['hotel_id'] = dim_hotel.index + 1
```

```python
dim_hotel.head()
```

```python
dim_guest = df[['name','email','phone-number','country','customer_type']].drop_duplicates().reset_index(drop=True)
dim_guest['guest_id'] = dim_guest.index + 1
dim_guest.head()
```

```python
dim_room = df[['reserved_room_type', 'assigned_room_type']].drop_duplicates().reset_index(drop=True)
dim_room['room_id'] = dim_room.index + 1
dim_room.head(100)
```

```python
dim_meal = df[['meal']].drop_duplicates().reset_index(drop=True)
dim_meal['meal_id'] = dim_meal.index + 1
dim_meal.head(10)
```

```python
dim_market = df[['market_segment', 'distribution_channel']].drop_duplicates().reset_index(drop=True)
dim_market['market_id'] = dim_market.index + 1
dim_market.head(100)
```

```python
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'].astype(str) + '-' +
    df['arrival_date_day_of_month'].astype(str),
    errors='coerce'
)

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

# Collect unique dates from BOTH arrival & reservation_status_date
all_dates = pd.concat([df['arrival_date'], df['reservation_status_date']]).dropna().drop_duplicates().sort_values().reset_index(drop=True)

# Create DimDate table
dim_date = pd.DataFrame({'date': all_dates})
dim_date['date_id'] = dim_date.index + 1
dim_date['year'] = dim_date['date'].dt.year
dim_date['month'] = dim_date['date'].dt.month
dim_date['day'] = dim_date['date'].dt.day
dim_date['quarter'] = dim_date['date'].dt.quarter
dim_date.head()
```

```python
fact_bookings = df[['booking_id','hotel','reserved_room_type','assigned_room_type',
                    'market_segment','distribution_channel','meal',
                    'name','email','phone-number','country','customer_type',
                    'arrival_date','reservation_status_date',
                    'lead_time','is_canceled',
                    'adults','children','babies',
                    'stays_in_weekend_nights','stays_in_week_nights','adr','deposit_type',
                    'days_in_waiting_list','required_car_parking_spaces','total_of_special_requests',
                    'is_repeated_guest','previous_cancellations','previous_bookings_not_canceled',
                    'reservation_status']]
fact_bookings.head()
```

```python
# Merge with other dims
fact_bookings = (fact_bookings
    .merge(dim_hotel, on='hotel', how='left')
    .merge(dim_room, on=['reserved_room_type','assigned_room_type'], how='left')
    .merge(dim_market, on=['market_segment','distribution_channel'], how='left')
    .merge(dim_meal, on='meal', how='left')
    .merge(dim_guest, on=['name','email','phone-number','country','customer_type'], how='left')
    # arrival date â†’ dim_date
    .merge(dim_date[['date','date_id']], left_on='arrival_date', right_on='date', how='left')
    .rename(columns={'date_id':'arrival_date_id'})
    .drop(columns=['date'])
    # reservation_status_date â†’ dim_date
    .merge(dim_date[['date','date_id']], left_on='reservation_status_date', right_on='date', how='left')
    .rename(columns={'date_id':'reservation_date_id'})
    .drop(columns=['date'])
)
```

```python
# Keep final structure
fact_bookings = fact_bookings[['booking_id','hotel_id','room_id','market_id',
                               'meal_id','guest_id',
                               'arrival_date_id','reservation_date_id',
                               'lead_time','is_canceled',
                               'adults','children','babies',
                               'stays_in_weekend_nights','stays_in_week_nights','adr','deposit_type',
                               'days_in_waiting_list','required_car_parking_spaces','total_of_special_requests',
                               'is_repeated_guest','previous_cancellations','previous_bookings_not_canceled',
                               'reservation_status']]
fact_bookings.head()
```

```python
fact_bookings.to_csv("FactBookings.csv", index=False)
```
